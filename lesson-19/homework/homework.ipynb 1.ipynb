```python
# Homework Assignment 2: Examining Customer Orders
#You have a dataset (task\customer_orders.csv) containing information about customer orders. The dataset has the following columns:
import pandas as pd

data = {
    "OrderID": [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008, 1009, 1010],
    "CustomerID": [501, 502, 503, 501, 504, 505, 506, 502, 507, 508],
    "Product": ["Laptop", "Mouse", "Keyboard", "Monitor", "Laptop", "Mousepad", "Monitor", "Keyboard", "Webcam", "Headset"],
    "Quantity": [3, 4, 5, 2, 22, 3, 15, 10, 20, 2],
    "Price": [1000.00, 25.00, 45.00, 200.00, 1000.00, 10.00, 200.00, 45.00, 70.00, 60.00]
}
df = pd.DataFrame(data)
print(df)


```

```python
df
#Group the data by CustomerID and filter out customers who have made less than 20 orders.
x=df.groupby(['CustomerID'])['Quantity'].sum().reset_index()
y=x[x['Quantity']<20]
y
result_df = df[df['CustomerID'].isin(y['CustomerID'])]
result_df

```

```python
df
#Identify customers who have ordered products with an average price per unit greater than $120.
x=df.groupby(['CustomerID'])['Price'].mean().reset_index()
y=x[x['Price']>120]
result=df[df['CustomerID'].isin(y['CustomerID'])]
result
```
